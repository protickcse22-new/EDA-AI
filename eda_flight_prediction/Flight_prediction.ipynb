```python
## Feature engineering process
final_df['Date_of_Journey'].str.split('/')[0]
```

```python
final_df['Day'] = final_df['Date_of_Journey'].str.split('/').str[0]
final_df['Month'] = final_df['Date_of_Journey'].str.split('/').str[1]
final_df['Year'] = final_df['Date_of_Journey'].str.split('/').str[2]
```

```python
final_df.head(4)
```

```python
## df['Day'] = df['Date_of_Journey'].apply(lambda x:x.split("/")[0])
```

```python
final_df['Day'] = final_df['Day'].astype(int)
final_df['Month'] = final_df['Month'].astype(int)
final_df['Year'] = final_df['Year'].astype(int)
```

```python
final_df.info()
```

```python
final_df.drop('Date_of_Journey',axis=1,inplace=True)
```

```python
final_df.head(1)
```

```python
final_df['Arrival_Time'] = final_df['Arrival_Time'].apply(lambda x:x.split(' ')[0])
```

```python
final_df.head(4)
```

```python
final_df.isnull().sum()
```

```python
plt.rcParams["figure.figsize"] = (16,8)
sns.heatmap(final_df.isnull(),cmap='viridis',cbar=False,yticklabels=False)
```

```python
final_df.head(2)
```

```python
final_df['Arrival_hour'] = final_df['Arrival_Time'].apply(lambda x:x.split(":")[0]).astype(int)
final_df['Arrival_min'] = final_df['Arrival_Time'].apply(lambda x:x.split(":")[1]).astype(int)
```

```python
final_df.info()
```

```python
final_df.drop('Arrival_Time',axis=1,inplace=True)
```

```python
final_df.head(3)
```

```python
final_df['Dep_hour'] = final_df['Dep_Time'].apply(lambda x:x.split(":")[0]).astype(int)
final_df['Dep_min'] = final_df['Dep_Time'].apply(lambda x:x.split(":")[1]).astype(int)
```

```python
final_df.drop('Dep_Time',axis=1,inplace=True)
```

```python
final_df.head(5)
```

```python
final_df['Total_Stops'].unique()
```

```python
final_df['Total_Stops']=final_df['Total_Stops'].map({'non-stop':0,'1 stop':1,'2 stops':2,'3 stops':3,'4 stops':4,'nan':1})
```
