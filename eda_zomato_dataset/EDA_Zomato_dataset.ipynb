### Merging two dataframe

```python
path = 'Zomatodataset/Country-Code.xlsx'
df_country = pd.read_excel(path)
```

```python
df_country
```

```python
final_df = pd.merge(df,df_country,on='Country Code',how='left')
```

```python
final_df.head(5)
```

```python
country_value = final_df.Country.value_counts().values
```

```python
country_Name = final_df.Country.value_counts().index
```

```python
country_Name
```

```python
plt.pie(country_value[:3],labels=country_Name[:3],autopct='%1.2f%%')
```

#### Observation: Zomato maximum records or transaction are from India after that USA and then UK

```python
final_df.columns
```

```python
ratings = final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Country Rating'})
```

```python
ratings
```

```python
ratings.dtypes
```

### Observation of rating
1. when rating is 4.9 to 4.5 -> Excellent
2. when rating is 4.4 to 4.0 -> Very Good
3. when rating is 3.9 to 3.5 -> Good
4. when rating is 3.4 to 2.5 -> Average
5. when rating is 2.4 to 1.8 -> Poor

Maximun number(2148) of people haven't given any rate.

```python
ratings.head()
```

```python
plt.rcParams['figure.figsize'] = (12,6)
sns.barplot(x="Aggregate rating",y="Country Rating",data=ratings)
```

```python
sns.barplot(x="Aggregate rating",y="Country Rating",data=ratings,hue='Rating color',palette=['blue','red','orange',
                                                                                            'yellow','green','green'])
```

### Observation
    1. Not rated count is very high
    2. Maximum number of rating between 2.5 to 3.4

```python
### count plot (for categorical values)
sns.countplot(x="Rating color",data=ratings,palette=['blue','red','orange','yellow','green','green'])
```
